````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df1=pd.read_csv('CarPrice_Assignment.csv')

df1['CarName']=df1['CarName'].astype('category')
df1['CarName']=df1['CarName'].cat.codes

df1['fueltype']=df1['fueltype'].astype('category')
df1['fueltype']=df1['fueltype'].cat.codes

df1['aspiration']=df1['aspiration'].astype('category')
df1['aspiration']=df1['aspiration'].cat.codes

df1['doornumber']=df1['doornumber'].astype('category')
df1['doornumber']=df1['doornumber'].cat.codes

df1['carbody']=df1['carbody'].astype('category')
df1['carbody']=df1['carbody'].cat.codes

df1['drivewheel']=df1['drivewheel'].astype('category')
df1['drivewheel']=df1['drivewheel'].cat.codes

df1['enginelocation']=df1['enginelocation'].astype('category')
df1['enginelocation']=df1['enginelocation'].cat.codes

df1['fuelsystem']=df1['fuelsystem'].astype('category')
df1['fuelsystem']=df1['fuelsystem'].cat.codes

df1['cylindernumber']=df1['cylindernumber'].astype('category')
df1['cylindernumber']=df1['cylindernumber'].cat.codes

df1['enginetype']=df1['enginetype'].astype('category')
df1['enginetype']=df1['enginetype'].cat.codes
df1.head()
````

````python
df1.isnull().sum()
````

````python
x=df1.drop(columns='price')
y=df1['price']

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x, y, test_size=0.3, random_state=0)

from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train, y_train)
c=lr.intercept_
m=lr.coef_

print("intercept is: ", c)
print("coefficient is: ", m)

from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
y_pred_train=lr.predict(x_train)
y_pred_test=lr.predict(x_test)

r2=r2_score(y_train, y_pred_train)
mse=mean_squared_error(y_train, y_pred_train)
mae=mean_absolute_error(y_train, y_pred_train)
print("r2_score is: ", r2)
print("mean_squared_error is: ", mse)
print("mean_abssolute_error is: ", mae)
````

````python
plt.scatter(y_train, y_pred_train, alpha=0.5)
plt.plot([y.min(), y.max()], [y.min(), y.max()])
plt.xlabel("Actual price")
plt.ylabel("Predicted price")
plt.title("Actual vs Predicted prices- Training")
plt.show()
````

````python
r2_train=r2_score(y_train, y_pred_train)
r2_test=r2_score(y_test, y_pred_test)
print("r2_score of training: ", r2_train)
print("r2 score of testing: ", r2_test)
````

````python
plt.scatter(y_test, y_pred_test, alpha=0.5)
plt.plot([y.min(), y.max()], [y.min(), y.max()])
plt.xlabel("Actual price")
plt.ylabel("Predicted price")
plt.title("Actual vs Predicted prices- Testing")
plt.show()
````

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('fastfood.csv')
df.head()
````

````python
df['restaurant']=df['restaurant'].astype('category')
df['restaurant']=df['restaurant'].cat.codes

df['item']=df['item'].astype('category')
df['item']=df['item'].cat.codes
````

````python
df.isnull().sum()
````

````python
df['fiber']=df['fiber'].fillna(value=4.0)
df['vit_a']=df['vit_a'].fillna(value=6.0)
df['vit_c']=df['vit_c'].fillna(value=20.0)
df['calcium']=df['calcium'].fillna(value=30.0)
````

````python
df.dropna(inplace=True)
````

````python
x=df.drop(columns='Healthy')
y=df['Healthy']

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x, y, test_size=0.3, random_state=21)

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)

from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(random_state=0).fit(x_train_scaled, y_train)

print("Predicted values of health (training dataset): \n", lr.predict(x_train_scaled)) # predicted values of Health (our dependant variable-y axis) (trained)
print("Score (training dataset): ", lr.score(x_train_scaled, y_train)) # 95% data is accurate (trained)

print("\n")

print("Predicted values of health (testing dataset): \n", lr.predict(x_test_scaled)) # predicted values of Health (our dependant variable-y axis) (tested)
print("Score (testing dataset): ", lr.score(x_test_scaled, y_test)) # 94% data is accurate (tested)
````

````python
# to improve the model:
lr1=LogisticRegression(random_state=0, C=1, fit_intercept=True).fit(x_train_scaled, y_train)
print("Predicted values of health (training dataset): \n", lr1.predict(x_train_scaled)) # predicted values of Health (our dependant variable-y axis) (trained)
print("Score (training dataset): ", lr1.score(x_train_scaled, y_train)) # 95% data is accurate (trained)

print("\n")

print("Predicted values of health (testing dataset): \n", lr1.predict(x_test_scaled)) # predicted values of Health (our dependant variable-y axis) (tested)
print("Score (testing dataset): ", lr1.score(x_test_scaled, y_test)) # 94% data is accurate (tested)
````
