```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
data=pd.read_csv('DelhiNCR Restaurants (1).csv')
```

Q1.Are there any null values in the dataset?if yes then fill them..

```python
data.isnull().sum() # to check null values
```

```python
data['Delivery_Rating']=data['Delivery_Rating'].fillna(value=4) # fills null values in each column
```

```python
data['Known_For2']=data['Known_For2'].fillna(value='Sizzling Brownies & Ice cream')
```

```python
data['Known_For22']=data['Known_For22'].fillna(value='Artistic Decor')
```

Q2.Which restaurant is the best one (having ratings more than 4.5 for delivery)
